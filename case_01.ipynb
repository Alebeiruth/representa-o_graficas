```python
# Primeiro, importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# MATPLOLIB

```python
# Dados de exemplo
anos = [2018, 2019, 2020, 2021, 2022]
vendas = [100, 150, 200, 250, 300]

# Criar o gráfico
plt.plot(anos, vendas, marker='o')
plt.title("Vendas por Ano")
plt.xlabel("Ano")
plt.ylabel("Vendas")
plt.grid(True)
plt.show()
```

```python
categorias = ['Tecnologia', 'Moda', 'Alimentos', 'Beleza']
valores = [400, 300, 500, 350]

plt.bar(categorias, valores, color='skyblue')
plt.title("Vendas por Categoria")
plt.xlabel("Categoria")
plt.ylabel("Valor")
plt.show()
```

```python
fatias = [400, 300, 500, 350]
labels = ['Tecnologia', 'Moda', 'Alimentos', 'Beleza']

plt.pie(fatias, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Distribuição de Vendas por Categoria")
plt.show()
```

# SEABORN

```python
# Criar um DataFrame de exemplo
data = {'idade': [18, 25, 35, 45, 55],
        'gastos': [200, 300, 500, 700, 800]}
df = pd.DataFrame(data)

# Gráfico DE DISPERSÃO
sns.scatterplot(x='idade', y='gastos', data=df)
plt.title("Disperção - Idade vs Gastos")
plt.show()
```

```python
# Dados de exemplo
data = np.random.rand(10, 3)
df = pd.DataFrame(data, columns=['A', 'B', 'C'])

# Matriz de correlação
corr = df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Matriz de Correlação")
plt.show()
```

# PANDAS

```python
# Criar uma tabela
tabela = pd.DataFrame({
    'Categoria': ['Tecnologia', 'Moda', 'Alimentos', 'Beleza'],
    'Vendas': [400, 300, 500, 350]
})

# Exibir a tabela
print(tabela)
```

```python
# Dados
tabela = [['Tecnologia', 400], ['Moda', 300], ['Alimentos', 500], ['Beleza', 350]]

# Criar a tabela no gráfico
fig, ax = plt.subplots()
ax.axis('tight')
ax.axis('off')
ax.table(cellText=tabela, colLabels=['Categoria', 'Vendas'], loc='center')
plt.show()
```

# PUXANDO OS DADOS DE UM CSV

```python
# Supondo que os dados estão em um arquivo chamado 'dados.csv' na pasta 'data'
file_path = "/mnt/data/dados.csv"

# Carregando o dataset
try:
    # Lendo os dados
    df = pd.read_csv(file_path)
    print("Dataset carregado com sucesso!\n")
    print(df.head())  # Exibe as primeiras linhas do dataset para referência
except FileNotFoundError:
    print(f"Arquivo '{file_path}' não encontrado. Por favor, verifique o caminho.")

# Criação de gráficos com Matplotlib, Seaborn e exibição de tabelas com Pandas

# ---------------------
# 1. Gráfico com Matplotlib (Exemplo: Gráfico de Barras)
if 'Categoria' in df.columns and 'Vendas' in df.columns:
    plt.figure(figsize=(8, 5))
    plt.bar(df['Categoria'], df['Vendas'], color='skyblue')
    plt.title("Vendas por Categoria")
    plt.xlabel("Categoria")
    plt.ylabel("Vendas")
    plt.show()
else:
    print("Colunas 'Categoria' e 'Vendas' não encontradas no dataset.")

# ---------------------
# 2. Gráfico com Seaborn (Exemplo: Gráfico de Dispersão)
if 'Idade' in df.columns and 'Gastos' in df.columns:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x='Idade', y='Gastos', data=df)
    plt.title("Idade vs Gastos")
    plt.xlabel("Idade")
    plt.ylabel("Gastos")
    plt.show()
else:
    print("Colunas 'Idade' e 'Gastos' não encontradas no dataset.")

# ---------------------
# 3. Exibição de tabela com Pandas
if 'Categoria' in df.columns and 'Vendas' in df.columns:
    print("\nTabela de Categorias e Vendas:")
    print(df[['Categoria', 'Vendas']])
else:
    print("Colunas 'Categoria' e 'Vendas' não encontradas para exibição de tabela.")
```
